# ðŸ¥‡ Top 10 countries


```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd

# read the data frame 
df = pd.read_csv('/kaggle/input/2021-olympics-medals-in-tokyo/Tokyo Medals 2021.csv')
df.head()
```

```python
# gen sorted data
df_rank = df.sort_values(by = ['Total'],ascending=False)
df_rank = df_rank[:11]
df_rank
```

```python
# opt-1: seaborn
plt.figure(figsize = (20,6))#change the plot size
plt.title("Top 10 countries with the highest tatal medals")
palette = sns.color_palette("OrRd",n_colors=11)
palette.reverse()
plot1 = sns.barplot(x="Total", y="Country", data=df_rank, palette = palette)
plt.show()
```
